```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
iris = pd.read_csv('Iris.csv')
iris.head(10)
```

```python
plt.scatter(iris['SepalLength'],iris['Species'])
```

```python
plt.scatter(iris['PetalLength'],iris['Species'])  
```

```python
plt.scatter(iris['SepalWidth'],iris['Species'])
```

```python
plt.scatter(iris['PetalWidth'],iris['Species'])
```

```python
x=iris[['SepalLength','PetalLength','SepalWidth','PetalWidth']]
y=iris['Species']
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
x_train
```

```python
x_test
```

```python
len(x_test)
```

```python
y_test
```

```python
from sklearn.linear_model import LogisticRegression
clf=LogisticRegression()
```

```python
clf.fit(x_train,y_train)
```

```python
cf=clf.predict(x_test)
print(cf)
```

```python
from sklearn.metrics import confusion_matrix
results = confusion_matrix(y_test, cf)
print(results)
```

```python
from sklearn.tree import DecisionTreeClassifier
c=DecisionTreeClassifier()
```

```python
c.fit(x_train,y_train)
```

```python
predict_decisionTrees=c.predict(x_test)
print(predict_decisionTrees)
```

```python
results = confusion_matrix(y_test,predict_decisionTrees)
print(results)
```

```python
from sklearn.neighbors import KNeighborsClassifier
alg=KNeighborsClassifier()
```

```python
alg.fit(x_train,y_train)
```

```python
predict_KNN=alg.predict(x_test)
print(predict_KNN)
```

```python
results = confusion_matrix(y_test,predict_KNN)
print(results)
```

```python
from sklearn.naive_bayes import GaussianNB
al=GaussianNB()
```

```python
al.fit(x_train,y_train)
```

```python
predict_naive=al.predict(x_test)
print(predict_naive)
```

```python
results = confusion_matrix(y_test,predict_naive)
print(results)
```

```python
from sklearn.svm import SVC
a=SVC()
```

```python
a.fit(x_train,y_train)
```

```python
predict_SVC=a.predict(x_test)
print(predict_SVC)
```

```python
results = confusion_matrix(y_test,predict_SVC)
print(results)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier()
```

```python
rf.fit(x_train,y_train)
```

```python
predict_forest=rf.predict(x_test)
print(predict_forest)
```

```python
results = confusion_matrix(y_test,predict_forest)
print(results)
```
